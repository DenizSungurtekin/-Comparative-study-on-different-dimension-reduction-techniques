```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
res = [[0.029677460463783056, ('perplexity = ', 1, 'mode = ', 'random')],
 [0.05535811539052515, ('perplexity = ', 15, 'mode = ', 'random')],
 [0.009997459316534063, ('perplexity = ', 1, 'mode = ', 'gaussian')],
 [0.0, ('perplexity = ', 15, 'mode = ', 'gaussian')],
 [0.05423976074028536, ('perplexity = ', 1, 'mode = ', 'laplace')],
 [0.03659586771124623, ('perplexity = ', 15, 'mode = ', 'laplace')],
 [0.9934273044746114, ('perplexity = ', 1, 'mode = ', 'pca')],
 [1.0, ('perplexity = ', 15, 'mode = ', 'pca')]]
```

```python
res = np.asarray(res)
print(res.shape)
```

```python
colors = ['b', 'c', 'y', 'm']
markers = ["o","x"]
y = res[:,0]
labels = res[:,1]
print(labels)
```

```python
x = [i for i in range(len(y))]
size = len(y)
indexs = [i for i in range(size)]
indexs2 = np.asarray([[i]*2 for i in range(int(size/2))])
indexs2 = indexs2.flatten()

for i in indexs:
    plt.scatter(x[i],y[i],color = colors[indexs2[i]],marker = markers[i%2],label=labels[i])


plt.legend(loc='upper left',ncol=2,fontsize=5)
plt.show()
```
